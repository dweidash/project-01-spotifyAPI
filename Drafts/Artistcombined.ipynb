```python
# Dependencies and Setup
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
artistscombined = "artistcombined.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
artistscombined = "artistcombined.csv"
artistdf = pd.read_csv(artistscombined)
```

```python
artistnumber = 0
# Display the data table for preview
print(artistdf)
artistnumber = len(artistdf) # length of artists
```

```python
# Clean column names and remove excess columns
artistdf = artistdf.drop('Unnamed: 0', 1) 
artistdf.rename(columns = {'2018':'Poll Ranking'}, inplace = True) 
artistdf.rename(columns = {'popularity':'Popularity Scale'}, inplace = True)
artistdf.rename(columns = {'Ticket_Sales': 'Ticket Sales'}, inplace = True)
artistdf.rename(columns = {'followers': 'Followers'}, inplace = True)
artistdf.rename(columns = {'genre': 'Genres'}, inplace = True)
artistdf[["Ticket Sales"]] = artistdf[["Ticket Sales"]].apply(pd.to_numeric)
# Display Dataframe
artistdf.head(10)
```

```python
colors = ('r')
plt.figure(figsize=(10, 10))
plt.scatter(artistdf["Followers"], artistdf["Ticket Sales"], c=colors, alpha=0.5, edgecolor="black") #Ticket Sales and Followers
plt.ylabel('Ticket Sales')
plt.yticks(np.arange(0, 5500000, step=100000))
plt.title('Ticket Sales vs Followers')
plt.xlabel('Followers')
plt.savefig('figure.png')
plt.show()
```
